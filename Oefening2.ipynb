# Interactieve Python-oefening
In deze oefening gaan we een klein interactief programma maken.

We gaan de gebruiker vragen om hun naam en leeftijd in te vullen en een gepersonaliseerde boodschap tonen.

**Wat leer je hier?**
- Hoe je `input()` gebruikt om gebruikersinput te krijgen
- Hoe je variabelen gebruikt om gegevens op te slaan
- Hoe je `print()` gebruikt om output te tonen

```python
# Vraag de gebruiker om zijn naam en leeftijd
naam = input("Wat is je naam? ")
leeftijd = input("Hoe oud ben je? ")

# Toon een gepersonaliseerde boodschap
print("Hallo " + naam + ", je bent " + leeftijd + " jaar oud!")
```

### Wat doet deze code?
- `input()` vraagt de gebruiker om iets in te vullen.
- `naam` en `leeftijd` slaan deze informatie op.
- `print()` combineert de antwoorden en toont een gepersonaliseerde boodschap.
